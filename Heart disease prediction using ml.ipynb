```python
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import confusion_matrix, accuracy_score, recall_score, precision_score, f1_score
from sklearn.preprocessing import normalize

# Load the iris dataset
iris = load_iris()
print(iris.data.shape)

# Separate the data from the target attributes
x = iris.data
y = iris.target

# Normalize the data attributes
normalize_x = normalize(x)

# Split the dataset into training and test sets
X_train, X_test, y_train, y_test = train_test_split(normalize_x, y, test_size=0.2, random_state=15)

# Train the Logistic Regression model
clf = LogisticRegression().fit(X_train, y_train)

# Make predictions
y_pred = clf.predict(X_test)

# Compute confusion matrix
cm = confusion_matrix(y_test, y_pred)
print("Confusion Matrix:")
print(cm)

# Compute accuracy score
accuracy = accuracy_score(y_test, y_pred)  # Fixed typo: 'y_predict' -> 'y_pred'
print("Accuracy Score:", accuracy)

# Compute other metrics
recall = recall_score(y_test, y_pred, average='macro')
precision = precision_score(y_test, y_pred, average='macro')
f1 = f1_score(y_test, y_pred, average='macro')

print("Recall Score:", recall)
print("Precision Score:", precision)
print("F1 Score:", f1)
```

```python
from sklearn.model_selection import GridSearchCV
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score

# Initialize the classifier
clf = LogisticRegression(max_iter=1000)  # Increase the number of iterations

grid_values = [
    # Grid for 'l1' penalty, only 'liblinear' solver can be used
    {
        'penalty': ['l1'],
        'C': [0.01, 1, 10],
        'solver': ['liblinear']
    },
    # Grid for 'l2' penalty, both solvers are supported
    {
        'penalty': ['l2'],
        'C': [0.01, 1, 10],
        'solver': ['liblinear', 'lbfgs']
    }
]

# Set up GridSearchCV
grid_clf_acc = GridSearchCV(clf, param_grid=grid_values, scoring='accuracy')

# Fit the model with the training data
grid_clf_acc.fit(X_train, y_train)

# Print the best parameters and best score from grid search
print("Best Parameters:", grid_clf_acc.best_params_)
print("Best Score:", grid_clf_acc.best_score_)

# Make predictions on the test set using the best model
y_pred = grid_clf_acc.predict(X_test)

# Calculate and print accuracy on the test set
accuracy = accuracy_score(y_test, y_pred)
print("Test Set Accuracy:", accuracy)
```
